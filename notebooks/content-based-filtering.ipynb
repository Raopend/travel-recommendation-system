```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/raopend/Workspace/travel-recommendation-system/notebooks/dataset/tours_truncated.csv")
df.head()
```

```python
df = df[['display_name', 'reviews_count', 'tags_desc', 'address', 'review_rating']]
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv.fit(df['tags_desc'])
cv.vocabulary_.keys()
```

```python
cv_matrix = cv.transform(df['tags_desc']) 
cv_matrix.shape
```

```python
cv_matrix.todense()
```

```python
pd.DataFrame(
    cv_matrix.todense(),
    columns=list(cv.vocabulary_.keys()),
    index=df['display_name']
).sample(5)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(cv_matrix)
cosine_sim
```

```python
cosine_sim_df = pd.DataFrame(cosine_sim,index=df['display_name'], columns=df['display_name'])
cosine_sim_df.sample(5,axis=1).sample(10,axis=0)
```

```python
def tourism_recommendations(place_name,similarity_data=cosine_sim_df,items=df[['display_name', 'tags_desc', 'address']], k=5):
    index = similarity_data.loc[:,place_name].to_numpy().argpartition(range(-1,-k,-1))
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    closest = closest.drop(place_name,errors='ignore')
    return pd.DataFrame(closest).merge(items).head(k)
tourism_recommendations("苏州古典园林")
```
